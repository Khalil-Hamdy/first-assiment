2: Sort following NumPy array
 Case 1: Sort array by the second row
 Case 2: Sort the array by the second column

```python


array = np.array([[34,43,73],[82,22,12],[53,94,66]]) 
print(array)

byRow = array[:,array[1,:].argsort()]
print("Sorting by row")
print(byRow)

print("Sorting by column")
byColumn = array[array[:,1].argsort()]
print(byColumn)
```

3: Print max from axis 0 and min from axis 1 from the
following 2-D array.

```python
array = np.array([[34,43,73],[82,22,12],[53,94,66]]) 
print (array)

minOfAxisOne = np.amin(array, 1) 
print('min of axis one : ')
print(minOfAxisOne)

maxOfAxisOne = np.amax(array, 0) 
print('max of axis one : ')
print(maxOfAxisOne)
```
